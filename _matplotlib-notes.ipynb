## 1. Pyplot Basics

### 1.1 Importing & Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

`%matplotlib inline`: Only for Jupyter Notebook. By default, it doesn't know whether you want it to display charts as you create them or not. This tells it to display charts.

### 1.2 Key Terms

**Figure**: The whole canvas.  
**Axes**: Chart area inside the canvas. (You can have multiple charts within a canvas)  
**Pyplot (plt)**: A state machine that keeps track of current Figure/Axes behind the scenes.

### 1.3 Quick Plotting

We'll plot using **Pyplot**. It's not the recommended way to plot, but it's a good way to quickly plot something.

```python
plt.plot([2, 4, 1, 7]) # Only y values, x is auto generated as index
plt.title("Demo")
plt.xlabel("index")
plt.ylabel("values")
plt.show()
```

Pyplot creates an anonymous figure and axes under the hood and applies these commands to them. It's considered a "quick" way because you don't create the figure and axes yourself which means you don't have full control over them.

**Code Explanation**
* `.plot()`: Adds the values.
* `.title()`: Sets the title of the axes.
* `.xlabel()`: Sets the label of the x-axis.
* `.ylabel()`: Sets the label of the y-axis.
* `.show()`: Outputs the resulting chart.

### 1.4 A Peak at Real Plotting

We'll plot using the **Object-Oriented API**. This is the recommended way, and the way that is commonly used.

```python
fig, ax = plt.subplots()
ax.plot([1, 3, 2], [4, 2, 7])
ax.set_title("Real Plotting")
plt.show()
```

With `fig, ax = plt.subplots()`, you manually create and define the Figure `fig` and the Axes `ax`, and now able to use their methods.

**Code Explanation**
* `fig, ax = plt.subplots()`: Manually creates and define the figure `fig` and the axes `ax`.
* `ax.plot()`: Adds values to the axes.
* `ax.set_title()`: Sets the title of the axes. Similar to `plt.title()`, but this time you choose which axes. (You can have more axes later)
* `plt.show()`: Outputs the resulting chart.

### 1.5 Quick Styling

```python
x = [0, 1, 2, 3, 4]
y = [2, 1, 3, 5, 4]
```

```python
plt.plot(x, y,
        marker="o",
        linestyle="-.",
        linewidth=2,
        alpha=0.5,
        label="Line Label",
        color='#07575b')
plt.legend()
plt.title("Styled Line")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

**Code Explanation**
* `marker`: Sets the style of the points on the line.
* `alpha`: Sets the opacity of the line.
* `label`: Sets the label of the line. Shows in the legend.
* `.legend()`: Shows the legend on the chart.

Just like with quick plotting vs real plotting, there is a better way to style.

### 1.6 Quick Saving

Since Pyplot doesn't seem to save the figure and axes when moving to a new cell. I will copy the previous plot and save it.

```python
# Previous Chart
plt.plot(x, y,
        marker="o",
        linestyle="-.",
        linewidth=2,
        alpha=0.5,
        label="Line Label",
        color='#07575b')
plt.legend()
plt.title("Styled Line")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Saving the Chart
plt.savefig("styled-chart.png", bbox_inches="tight", dpi=150)
```

`bbox_inches`: By default, Matplotlib saves the file with margin around it. Setting `bbox_inches="tight"` removes that margin.  
`dpi`: The higher, the higher the quality and image size.

## 2. Object-Oriented API

This is the recommended way of doing things. You'll be using this most of the time.

### 2.1 Plotting

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3], [4, 2, 7])

ax.set_title("OO Way Plotting")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

plt.show()
```

Figure `fig`: Whole canvas.  
Axes `ax`: One plotting area.

### 2.2 Multiple Lines

```python
y2 = [3, 4, 3, 1, 5]
```

```python
fig, ax = plt.subplots();

ax.plot(x, y, label="Series A")
ax.plot(x, y2, label="Series B")

ax.set_title("Multiple Lines Chart")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.legend()

plt.show()
```

### 2.3 Calculated Axis

```python
import numpy as np
```

```python
fig, ax = plt.subplots()

ax.plot(x, np.pow(x, x))

ax.set_title("x^x")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, np.sin(x))

ax.set_title("Sin(x)")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

plt.show()
```

### 2.4 Styling

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(x, y, label="Series A", linewidth=2, alpha=0.6, color="#07575b", linestyle="--", marker="o")
ax.plot(x, y2, label="Series B", linewidth=4, alpha=0.9, color="#f67280", linestyle="-.", marker="8")

ax.set_title("Styling Chart")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.legend()
ax.grid(True)
ax.set_xlim(0, 7)
ax.set_ylim(0, 9)

plt.show()
```

**More Methods and Parameters**
* `figsize()`: Sets the width and height of the figure/canvas.
* `.grid()`: Enables grid.
* `.set_xlim()`: Sets the start and end of the x-axis.
* `.set_ylim()`: Sets the start and end of the y-axis.

## 3. Types of Charts

```python
import numpy as np

a = np.random.randint(0, 100, size=1000)
b = np.random.randint(0, 100, size=1000)
```

## 3.1 Histogram `hist()`

Looks like a bar chart. Splits data into classes then counts data's frequency per class.

```python
fig, ax = plt.subplots()

ax.hist(a, bins=20, edgecolor="#61a4ad", color="#07575b")
ax.set_title("Histogram")
ax.set_xlabel("Value (In Classes)")
ax.set_ylabel("Frequency")


plt.show()
```

You can also overlay and compare two histograms.

```python
fig, ax = plt.subplots()

ax.hist(a, bins=20, edgecolor="#61a4ad", color="#07575b", label="Series A")
ax.hist(b, bins=20, edgecolor="#f67280", color="#c06c84", label="Series B")
ax.set_title("Two Overlayed Histograms")
ax.set_xlabel("Value (In Class)")
ax.set_ylabel("Frequency")
ax.legend()

plt.show()
```

### 3.2 Horizontal Bars `barh()`

Use horizontal bars when labels are long or want to emphasize ranking.

```python
names = ["Yazeed Ahmad Moh'd Al-Momani", "Yazeed A. Al-Momani", "Yazeed"]
score = [8.7, 4.2, 9.1]
```

```python
fig, ax = plt.subplots(figsize=(6,3))

ax.barh(names, score, height=0.5, color="#c06c84", edgecolor="#f67280")
ax.set_title("Horizontal Bar")
ax.set_xlabel("Score")
ax.set_ylabel("Names")

plt.show()
```

### 3.3 Vertical Bars `bar()`

Great for comparing categories.

```python
fig, ax = plt.subplots()

ax.bar(letters, values1, color="#07575b", edgecolor="#61a4ad", width=0.6)
ax.set_title("Vertical Bars")
ax.set_xlabel("Category (Letters)")
ax.set_ylabel("Values")

plt.show()
```

### 3.4 Line `plot()`

Same one we are used to.

### 3.5 Scatter `scatter()`

Points on the axes.

```python
a = np.random.randint(0, 100, 500)
b = np.random.randint(0, 100, 500)
c = np.random.randint(0, 100, 350)
d = np.random.randint(0, 100, 350)
e = np.random.randint(0, 2, 350)
```

```python
fig, ax = plt.subplots()

ax.scatter(a, b, s=40, color="#07575b")
ax.set_title("Scatter Chart")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-label")

plt.show()
```

`s`: point size.

Overlay with another scatter:

```python
fig, ax = plt.subplots()

ax.scatter(a, b, s=40, label="Series AB", color="#07575b")
ax.scatter(c, d, s=40, label="Series CD", color="#c06c84")
ax.set_title("Overlay Scatter Chart")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.legend()

plt.show()
```

#### 3.5.1 Colormap

Coloring by a third variable.

```python
fig, ax = plt.subplots()

scatter = ax.scatter(c, d, c=e, cmap="coolwarm", s=40)
colorbar = fig.colorbar(scatter, ax=ax)
colorbar.set_label("Temperature")

ax.set_title("Scatter Chart")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-label")

plt.show()
```

## 4. Axes

### 4.1 Using Multiple Axes

#### 4.1.1 Creating Multiple Axes

```python
fig, ax = plt.subplots(2, 2, figsize=(6,6))

ax[0,0].plot(x, y, color="#07575b", marker="8", linestyle="--", linewidth=2)

ax[0,1].scatter(a, b, s=4, label="Series AB", color="#07575b")

ax[1,0].bar(letters, values1, color="#07575b", edgecolor="#61a4ad", width=0.6)

ax[1,1].plot(y, x, color="#07575b", marker="8", linestyle="--", linewidth=2)

fig.suptitle("Multiple Axes Figure")

plt.show()
```

`fig.suptitle()`: Sets the title for the entire figure.

#### 4.1.2 Shared Axes

```python
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig, axs = plt.subplots(1, 2, sharey=True)

axs[0].plot(x, y1, color="#c06c84", linestyle="--", linewidth=2, marker="8")
axs[0].set_title("sin")

axs[1].plot(x, y2, color="#07575b", linestyle="--", linewidth=2, marker="o")
axs[1].set_title("cos")

plt.show()
```

When all axes in a figure share the same y-axis or x-axis, you can set a `sharey=True` or `sharex=True` in the `subplots()` to share the axis between them. Which means it only writes the axis once, making it more readable. Also, you need to pass in the same y or x for all axes.

### 4.2 More Styling Methods

#### 4.2.1 Reference Lines

```python
x = [0, 1, 2, 3, 4]
y = [2, 1, 3, 5, 4]
```

```python
fig, ax = plt.subplots()

ax.plot(x, y, color="#07575b", marker="8", linestyle="--", linewidth=2)
ax.set_title("Line Chart")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

# Reference Lines
ax.axvline(2.5, color="#843b62", linestyle="--", linewidth=2)
ax.axhline(2.5, color="#843b62", linestyle="--", linewidth=2)

plt.show()
```

Of course, you can place as much and as little vertical and horizantal reference lines as you want.  

You can also use calculated numbers instead of constant numbers, like mean and median.

```python
fig, ax = plt.subplots()

ax.plot(x, y, color="#07575b", marker="8", linestyle="--", linewidth=2)
ax.set_title("Line Chart")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

# Reference Line
ax.axhline(np.median(y), color="#843b62", linestyle=":", linewidth=2)
ax.text(0, np.median(y)+0.1, "Median Line", color="#843b62")

plt.show()
```

`ax.text()`: Adds a text on the axes. The first two parameters are the coordinates on the axes (x-value and y-value), the third parameter is the text itself.

#### 4.2.2 Grid Styling

```python
fig, ax = plt.subplots()

ax.plot(x, y, color="#07575b", marker="8", linestyle="--", linewidth=2)
ax.set_title("Line Chart")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

# Grid
ax.grid(visible=True, alpha=0.4, linestyle="-.")

plt.show()
```
